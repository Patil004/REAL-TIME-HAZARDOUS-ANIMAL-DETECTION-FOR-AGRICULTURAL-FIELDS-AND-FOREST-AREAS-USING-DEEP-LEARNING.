```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# Load model
model = YOLO("/kaggle/input/models/ankitpatil04/animal-detectaion-github/other/default/1/best (6).pt")

# Define hazardous animals
hazardous_animals = ["elephant", "bear", "tiger", "lion", "leopard", "boar", "pig"]

# Image path
img_path = "/kaggle/input/datasets/ankitpatil04/pig-image/domestic-pigs-sus-scrofa-domestica.webp"

# Read image
image = cv2.imread(img_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Run detection
results = model(img_path)

# Draw boxes
for r in results:
    for box in r.boxes:
        cls = int(box.cls[0])
        label = model.names[cls]
        conf = float(box.conf[0])

        x1, y1, x2, y2 = map(int, box.xyxy[0])

        # Check hazardous
        if label in hazardous_animals:
            color = (255, 0, 0)  # RED
            text = f"{label} (Hazard)"
        else:
            color = (0, 255, 0)  # GREEN
            text = f"{label} (Normal)"

        # Draw rectangle
        cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)

        # Put label
        cv2.putText(image, text, (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

# Show image in Kaggle
plt.figure(figsize=(10,10))
plt.imshow(image)
plt.axis("off")
plt.show()
```
